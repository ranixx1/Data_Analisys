```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)

# Verificando resultado
df.head()
```

```python
df['senioridade'].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
renomear_categorias_senioridade = {
    'senioridade': {
        'MI': 'Pleno',
        'SE': 'Sênior',
        'EN': 'Júnior',
        'EX': 'Executivo'
    },
}

df['senioridade']=df['senioridade'].replace(renomear_categorias_senioridade['senioridade'])
df['senioridade'].value_counts()
```

```python
renomear_categorias_contrato = {
    'contrato': {
        'FT': 'Tempo Integral',
        'PT': 'Meio-Tempo',
        'FL': 'Híbrido',
        'CT': 'Contrato'
    },
}

df['contrato']=df['contrato'].replace(renomear_categorias_contrato['contrato'])
df['contrato'].value_counts()
```

```python
renomear_categorias_tamanho_empresa= {
    'tamanho_empresa': {
        'S': 'Pequeno',
        'M': 'Médio',
        'L': 'Grande'
    },
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(renomear_categorias_tamanho_empresa['tamanho_empresa'])
df['tamanho_empresa'].value_counts()
```

```python
renomear_categorias_remoto = {
    'remoto': {
        0: 'Presencial',
        50: 'hibrido',
        100: 'Home oficce'
    },
}
df['remoto'] = df['remoto'].replace(renomear_categorias_remoto['remoto'])
df['remoto'].value_counts()
```
