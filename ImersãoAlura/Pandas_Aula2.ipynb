```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)

# Verificando resultado
df.head()
```

```python
renomear_categorias_senioridade = {
    'senioridade': {
        'MI': 'Pleno',
        'SE': 'Sênior',
        'EN': 'Júnior',
        'EX': 'Executivo'
    },
}

df['senioridade']=df['senioridade'].replace(renomear_categorias_senioridade['senioridade'])
df['senioridade'].value_counts()
```

```python
renomear_categorias_contrato = {
    'contrato': {
        'FT': 'Tempo Integral',
        'PT': 'Meio-Tempo',
        'FL': 'Híbrido',
        'CT': 'Contrato'
    },
}

df['contrato']=df['contrato'].replace(renomear_categorias_contrato['contrato'])
df['contrato'].value_counts()
```

```python
renomear_categorias_tamanho_empresa= {
    'tamanho_empresa': {
        'S': 'Pequeno',
        'M': 'Médio',
        'L': 'Grande'
    },
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(renomear_categorias_tamanho_empresa['tamanho_empresa'])
df['tamanho_empresa'].value_counts()
```

```python
renomear_categorias_remoto = {
    'remoto': {
        0: 'Presencial',
        50: 'hibrido',
        100: 'Home oficce'
    },
}
df['remoto'] = df['remoto'].replace(renomear_categorias_remoto['remoto'])
df['remoto'].value_counts()
```

```python
import numpy as np

#Cria DataFrame de teste
df_salarios = pd.DataFrame({
    'nome':['Ranilton','André','Ramon','Agenor','Geraldo'],
    'salario':[4000,np.nan,5000,np.nan,100000]
})
# Calcula a média salarial e substitui os nulos pela média
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# Cacula a mediana e substitui os valores null
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['segunda','Terça','Quarta','Quinta','Sexta','Sabádo','Domingo'],
    'Temperatura': [30,30,np.nan,np.nan,28,np.nan,27]
})
df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome':['Ranilton','André','Ramon','Agenor','Geraldo'],
    'cidade':['São Paulo',np.nan,'Rio de Janeiro',np.nan,'Parnamirim']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```
